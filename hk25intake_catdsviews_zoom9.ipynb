# Catalogs, expanded for convenient view 

```python
import healpix
import xarray as xr
import intake
# catalogue from which node? 
cat = intake.open_catalog("https://digital-earths-global-hackathon.github.io/catalog/catalog.yaml")["NCAR"]
```

```python
list(cat)
```

```python
ZOOMLEVEL = 9 

datasets = list(cat)
xr.set_options(display_expand_attrs = False, display_expand_coords = False)

for dsname in datasets: 
    print('********************* ',dsname,' ******************')
    try: 
        display( cat[dsname](zoom=ZOOMLEVEL).to_dask() )
    except: 
        print("No zoom level ", ZOOMLEVEL)
```
